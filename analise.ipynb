```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Etapa 1 – Tratamento e preparação dos dados

```python
base = pd.read_excel('Base Vendas.xlsx')
base.head(5)
```

### EDA 

```python
print(base.shape)
```

```python
base.info()
```

```python
# Remover colunas completamente vazias
base = base.dropna(axis=1, how='all')
print(base.columns)
```

```python
# Remover espaços e padronizar nomes
base.columns = base.columns.str.strip().str.replace(" ", "_")
base.head()
```

### Tratar valores ausentes (NaN)

```python
# Remover a coluna 'Unnamed: 11'
base = base.drop(columns=['Unnamed:_11'])

base.info()
```

```python
base = base.drop_duplicates()

# Criar colunas auxiliares de análise
base['Receita'] = base['Quantidade_Vendida'] * base['Preco_Unitario']
base['Custo_Total'] = base['Quantidade_Vendida'] * base['Custo_Unitario']
base['Lucro'] = base['Receita'] - base['Custo_Total']
base['Margem_%'] = round(base['Lucro'] / base['Receita'] * 100, 2)
```

```python
# Criar coluna de idade dos clientes
base_ref = pd.Timestamp('2025-12-31')
base['Idade'] = (base_ref - base['Data_de_Nascimento']).dt.days // 365

# Criar colunas de ano e mês da venda

base['Ano'] = base['Data_da_Venda'].dt.year
base['Mes'] = base['Data_da_Venda'].dt.month_name(locale='pt_BR')

base.head()
```

### Perguntas de Negócio (com análises)

Qual loja teve o maior faturamento total?

```python
vendas_por_loja = base.groupby('Loja')['Receita'].sum().sort_values(ascending=False)
print(vendas_por_loja)

"""ax =vendas_por_loja.plot(kind='bar', title='Faturamento Total, por Loja', xlabel='Loja', ylabel='Receita (R$)')
ax.bar_label(ax.containers[0], fmt='{:,.0f}', fontsize=7, label_type='center')

plt.tight_layout()
plt.show()"""""
import plotly.express as px

ax = vendas_por_loja.reset_index(name='Receita')

fig = px.bar(ax, x='Loja', y='Receita', title='Faturamrnto Total por Loja Interativo')

fig.update_traces(text=ax['Receita'].map('R$ {:,.0f}'.format), 
                  textposition='outside',
                  textfont_color='blue',
                  textfont_size=11)

fig.show()
```

Quais produtos são mais lucrativos?

```python
lucros_produtos = base.groupby('Produto')['Lucro'].sum().sort_values(ascending=False)
print(lucros_produtos)

ax = lucros_produtos.reset_index()

fig = px.bar(ax, x="Produto", y="Lucro", title='Lucros Totais por Produtos',
           labels={'Produto': 'Produto', 'Lucro': 'Lucro (R$)'},
           hover_data={'Lucro': ':.2f'})

fig.update_traces(text=ax['Lucro'].map('R$ {:,.0f}'.format), 
                  textposition='outside',
                  textfont_color='blue',
                  textfont_size=11)

fig.show()
```

Como as vendas evoluíram ao longo do tempo (por mês)?

```python
import plotly.express as px

# Agrupar e preparar os dados (igual ao seu código original)
vendas_mensais = base.groupby(['Ano', 'Mes'])['Receita'].sum().reset_index()
vendas_mensais.sort_values(by=['Ano', 'Mes'], inplace=True)

# Criar o gráfico interativo com Plotly
fig = px.line(
    vendas_mensais,
    x='Mes',
    y='Receita',
    color='Ano',  # diferencia as linhas por ano
    markers=True,
    title='📈 Evolução das Vendas Mensais'
)

# Personalizações opcionais
fig.update_layout(
    xaxis_title='Mês',
    yaxis_title='Receita Total',
    hovermode='x unified',
    template='plotly_white'
)

fig.update_traces(
    mode='lines+markers+text',
    text=vendas_mensais['Receita'],
    textposition='top center',
    texttemplate='%{text:,.0f}'  # formata os números com milhar (ex: 1.200)
)

# Mostrar o gráfico
fig.show()
```

Qual é o ticket médio (valor médio por venda)?

```python


ticket_medio = base['Receita'].mean()
print(f"Ticket médio: R$ {ticket_medio:.2f}")
```

Qual o perfil etário dos clientes?

```python
faixa_etaria = pd.cut(base['Idade'], bins=[0,18,30,45,60,120],
                      labels=['Até 18','19-30','31-45','46-60','60+'])
```

```python
# Dados já calculados
idade_media_produto = base.groupby(faixa_etaria)['Receita'].mean().reset_index()
idade_media_produto.columns = ['Faixa Etária', 'Receita Média']

# Gráfico de pizza
fig = px.pie(
    idade_media_produto,
    names='Faixa Etária',
    values='Receita Média',
    title='Faixa Etária e Receita Média',
)

# Mostrar label + porcentagem dentro da fatia
fig.update_traces(
    textposition='inside',
    textinfo='percent+label'
)

fig.update_layout(template='plotly_white')

fig.show()
```

Quais produtos têm maior margem de lucro (%)

```python
margem = base.groupby('Produto')['Margem_%'].mean().sort_values(ascending=False)
print(margem)
```

```python
# Top 10 produtos pela margem média
margem = base.groupby('Produto')['Margem_%'].mean().sort_values(ascending=False).head(10)

# Transformar em DataFrame para o Plotly
margem_df = margem.reset_index()
margem_df.columns = ['Produto', 'Margem (%)']

# Gráfico de rosca
fig = px.pie(
    margem_df,
    names='Produto',
    values='Margem (%)',
    title='Top 10 Produtos com Maior Margem (%)',
    hole=0.5   # ← transforma em rosca
)

# Exibir porcentagens e rótulos
fig.update_traces(
    textposition='inside',
    textinfo='percent+label'
)

fig.update_layout(template='plotly_white')

fig.show()
```

Qual o mês mais lucrativo de cada ano?

```python
lucro_mensal = base.groupby(['Ano', 'Mes'])['Lucro'].sum().reset_index()
lucro_max = lucro_mensal.loc[lucro_mensal.groupby('Ano')['Lucro'].idxmax()]
print(lucro_max)
```
